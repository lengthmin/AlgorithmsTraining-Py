枚举的基本思想:有序的尝试每一种可能

asd + fgh = jkl  
使用数字 1 - 9,每个数字只能使用一次  
用python中的集合,来确保每个数字只出现一次

```python
for a in range(1,10):
    for s in range(1,10):
        for d in range(1,10):
            for f in range(1,10):
                for g in range(1,10):
                    for h in range(1,10):
                        for j in range(1,10):
                            for k in range(1,10):
                                for l in range(1,10):
                                    if(len(set([a,s,d,f,g,h,j,k,l])) == 9):
                                        if(a * 100 + s * 10 + d + f * 100 + g * 10 + h == j * 100 + k * 10 + l):
                                            print("%d%d%d + %d%d%d = %d%d%d"%(a,s,d,f,g,h,j,k,l))
```
