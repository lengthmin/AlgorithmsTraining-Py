```python
from copy import deepcopy
from random import randrange

n_step = ((0, 1), (1, 0), (0, -1), (-1, 0))
def gen_array(map_str: str) -> list:
    return [list(map(int, line)) for line in map_str.strip().split()]
maze1 = """
0000000000
0000000000
0000000000
0000000000
0000000000
0000010000
0000000000
0000000000
0000000000
0000000000
"""
maze2 = """
0010
0000
0010
0100
0001
"""
maze = gen_array(maze2)
# 地图大小
m, n = len(maze), len(maze[0])
print(f"地图大小：({m}, {n})")
```

```python
# 下一步尝试 顺序：右下左上

def dfs(x, y, step):
    global min_step
    global founded
#     if founded:
#         return
    if x - p == 0 and y - q == 0:
        if step < min_step: min_step = step 
        print("找到")
        print(routes)
        print(f"step: {step}")
        print("----------------")
        founded = True
        return
    
    for i in range(4):
        tx = x + n_step[i][0]
        ty = y + n_step[i][1]
        # 判断越界
        if tx < 0 or tx >= m or ty < 0 or ty >= n:
            continue
        print(f"(tx, ty): ({tx}, {ty})")
        print(f"maze[tx][ty]: {maze[tx][ty]}")
        print(f"book[tx][ty]: {book[tx][ty]}")
        print(f"step: {step}")
        print("==============")
        if maze[tx][ty] == 0 and book[tx][ty] == 0:
            book[tx][ty] = 1
            routes.append((tx, ty))
            dfs(tx, ty, step+1)
            routes.pop()
            book[tx][ty] = 0
    return
book = [ [0]*n for j in range(m) ]
start_x, start_y = 0, 0
print(f"起点: ({start_x}, {start_y})")
book[start_x][start_y] = 1

p, q = 3, 2
# randrange(start_x + 1, m), randrange(start_y + 1, n)
print(f"终点: ({p}, {q})")
min_step = 999
routes = []
founded = False
dfs(start_x, start_y, 0)
print(min_step)
```
