```python
a = [22,11,5,88,15,36,54,1555,4,65]
#冒泡排序
#我们拍集体照的时候,相邻的两个比较换来换去,就是冒泡排序
#每次比较两个相邻的元素,如果顺序错误就把它们换过来.
for i in range(len(a) - 1):
    #一共要换 len(a) - 1 趟
    for j in range(len(a)- 1 - i):
        
        if(a[j] < a[j+1]):
            #python两个数的交换就是可以这么简单粗暴
            a[j],a[j+1] = a[j+1],a[j]
            print(a)
```
